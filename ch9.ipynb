## Using Clustering for Preprocessing

```python
from sklearn.datasets import load_digits

X_digits, y_digits = load_digits(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_digits, y_digits)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
log_reg.score(X_test, y_test)
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
        ("kmeans", KMeans(n_clusters=50)),
        ("log_reg", LogisticRegression())
    ])
pipeline.fit(X_train, y_train)
```

```python
pipeline.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = dict(kmeans__n_clusters=range(98,100))
grid_clf = GridSearchCV(pipeline,param_grid, cv=3, verbose=2)
grid_clf.fit(X_train, y_train)
```

```python
grid_clf.best_params_
```

```python
grid_clf.score(X_test, y_test)
```

## Using Clustering for Semi-Supervised Learning

```python
n_labeled = 50
log_reg = LogisticRegression()
log_reg.fit(X_train[:n_labeled], y_train[:n_labeled])
```

```python
log_reg.score(X_test, y_test)
```

```python
# Find the representative images
k = 50
kmeans = KMeans(n_clusters=k)
X_digits_dist = kmeans.fit_transform(X_train)
representative_digit_idx = np.argmin(X_digits_dist, axis=0)
X_representative_digits = X_train[representative_digit_idx]
```

```python
plt.figure(figsize=(8, 2))
for index, X_representative_digit in enumerate(X_representative_digits):
    plt.subplot(k // 10, 10, index + 1)
    plt.imshow(X_representative_digit.reshape(8, 8), cmap="binary", interpolation="bilinear")
    plt.axis('off')

# save_fig("representative_images_diagram", tight_layout=False)
plt.show()
```

```python
y_train[representative_digit_idx] # <-- Note: This is cheating~
```

```python
# y_representative_digits = np.array([6, 8, 2, 0, 9, 2, 4, 4, 1, 4, 3, 7, 5, 5, 2, 6, 3, 1, 9, 2, 7, 5,
#        8, 3, 7, 5, 1, 8, 8, 9, 9, 9, 3, 7, 0, 2, 2, 8, 1, 1, 4, 5, 7, 4,
#        6, 5, 8, 1, 4, 0]) # <-- Manually labelling

y_representative_digits = y_train[representative_digit_idx] # <-- Note: This is cheating~
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_representative_digits, y_representative_digits)
log_reg.score(X_test, y_test)
```

```python
# Propagate to all instances in the same cluster
y_train_propagated = np.empty(len(X_train), dtype=np.int32)
for i in range(k):
    y_train_propagated[kmeans.labels_==i] = y_representative_digits[i]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train_propagated)
log_reg.score(X_test, y_test)
```

```python
# Propagate labels to 50% closest instances in the same cluster
percentile_closest = 50 # 20

X_cluster_dist = X_digits_dist[np.arange(len(X_train)), kmeans.labels_]
for i in range(k):
    in_cluster = (kmeans.labels_ == i)
    cluster_dist = X_cluster_dist[in_cluster]
    cutoff_distance = np.percentile(cluster_dist, percentile_closest)
    above_cutoff = (X_cluster_dist > cutoff_distance)
    X_cluster_dist[in_cluster & above_cutoff] = -1
    
partially_propagated = (X_cluster_dist != -1)
X_train_partially_propagated = X_train[partially_propagated]
y_train_partially_propagated = y_train_propagated[partially_propagated]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train_partially_propagated, y_train_partially_propagated)
log_reg.score(X_test, y_test)
```

```python
np.mean(y_train_partially_propagated == y_train[partially_propagated])
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000, noise=0.05)
# dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan = DBSCAN(eps=0.1, min_samples=5)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
len(dbscan.core_sample_indices_)
```

```python
dbscan.core_sample_indices_
```

```python
dbscan.components_
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=50)
X_dbscan_core_sample = dbscan.components_
y_dbscan_core_sample = dbscan.labels_[dbscan.core_sample_indices_]
knn.fit(X_dbscan_core_sample, y_dbscan_core_sample)
```

```python
X_new = np.array([[-0.5, 0], [0, 0.5], [1, -0.1], [2,1]])
knn.predict(X_new)
```

```python
knn.predict_proba(X_new)
```

```python
y_dist, y_pred_idx = knn.kneighbors(X_new, n_neighbors=1)
y_pred = y_dbscan_core_sample[y_pred_idx]
print(y_dist)
print(y_pred_idx)
print(y_pred.ravel())
# Overwrite the far instances to -1
y_pred[y_dist > 0.2] = -1
print(y_pred.ravel())
```
